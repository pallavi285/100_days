# Linear Search 
Check each element one by one until the target is found or the list ends.

```python
def linear_search(arr,target):
    for i in range(len(arr)): #go throught each index
        if arr[i]==target:    #checking if element matches
            return i          #return index if found
    return -1 #not found
```

## Example Usage

```python
numbers =[10,23,30,45,50]
x = 30 
result = linear_search(numbers,x)

if result !=-1:
    print("Element found at index",result)
else:
    print("Element not found")
```
