# Reverse the array 

```python
# Temp array
def reverse_array(arr):
    n = len(arr)
    temp = [0]*n
    
    for i in range(n):
        temp[i]=arr[n-i-1]    
    for i in range(n):
        arr[i] = temp[i]    
        
if __name__ == "__main__":
    
    arr = [1, 4, 3, 2, 6, 5]
    reverse_array(arr)
    for i in range(len(arr)):
        print(arr[i],end=' ')
```

```python
# use two pointer

def reverse_array(arr):
    n = len(arr)
    left = 0 
    right = n-1
    
    while left < right:
        arr[left],arr[right] = arr[right],arr[left]
        left+=1
        right-=1  
if __name__ == "__main__":
    
    arr = [1, 4, 3, 2, 6, 5]
    reverse_array(arr)
    for i in range(len(arr)):
        print(arr[i],end=' ')
```

```python
# swapping approch
def reverse_array(arr):
    n = len(arr)
    for i in range(n//2):
        temp= arr[i]
        arr[i] = arr[n-i-1]
        arr[n-i-1] = temp
    
if __name__ == "__main__":
    
    arr = [1, 4, 3, 2, 6, 5]
    reverse_array(arr)
    for i in range(len(arr)):
        print(arr[i],end=' ')
```

```python
def reverse_array(arr):
    arr.reverse()
    
if __name__ == "__main__":
    arr = [1, 4, 3, 2, 6, 5]
    reverse_array(arr)
    for i in range(len(arr)):
        print(arr[i],end=' ')
```
