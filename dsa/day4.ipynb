# reverse depend on K

```python
def reverse_in_groups(arr,k):
    i = 0
    n = len(arr)
    
    while i < n:
        left = i
        right = min(i+k-1,n-1)
        while left<right:
            arr[left],arr[right]= arr[right],arr[left]
            left +=1
            right -=1
            
        i+=k
    return arr
if __name__ == "__main__":
    k=3
    arr = [1, 4, 3, 2, 6, 5]
    print(reverse_in_groups(arr,k))
        
```
