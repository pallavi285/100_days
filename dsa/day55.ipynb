# Find Third largest element 

```python
def thirdLargest(arr):
    n = len(arr)
    arr.sort()
    return arr[n-3]
arr = [1, 14, 2, 16, 10, 20]
print(thirdLargest(arr))
```

```python
# Three loops 
def thirdLargest(arr):
    n = len(arr)
    first = second = third = float('-inf')
    for i in range(n):
        if arr[i]>first:
            first = arr[i]
    for i in range(n):
        if arr[i]>second and arr[i]<first:
            second = arr[i]
    for i in range(n):
        if arr[i]>third and arr[i]<second:
            third = arr[i]
    return third

arr = [1, 14, 2, 16, 10, 20]
print(thirdLargest(arr))
```
