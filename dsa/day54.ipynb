# find second largest in array 


```python
def getSecondLargest(arr):
    n = len(arr)
    arr.sort()
    for i in range(n-2,-1,-1):
        if arr[n-2]!=arr[n-1]:
            return arr[i]
    return -1

if __name__ == "__main__":
    arr = [12, 35, 1, 10, 34, 1]
    print(getSecondLargest(arr))
```

```python
# Two pass approch

def getSecondLargest(arr):
    n = len(arr)
    largest = secondLargest =-1
    # finding the largest element 
    for i in range(n):
        if arr[i]> largest:
            largest = arr[i]
            
    for i in range(n):
        if arr[i]>secondLargest and arr[i]!=largest:
            secondLargest = arr[i]
            
    return secondLargest

if __name__ == "__main__":
    arr = [12,35,1,56,1]
    print(getSecondLargest(arr))
```

```python
# one pass approch

def getSecondLargest(arr):
    n = len(arr)
    largest = secondLargest =-1
    # finding the largest element 
    for i in range(n):
        if arr[i]> largest :
            secondLargest = largest
            largest = arr[i]
        elif arr[i]<largest and arr[i]>secondLargest:
            secondLargest = arr[i]
            

            
    return secondLargest

if __name__ == "__main__":
    arr = [12,35,1,56,1]
    print(getSecondLargest(arr))
```

```python
def maxProduct(arr):
    n = len(arr)
    arr.sort()
    return max(arr[0] * arr[1] * arr[n - 1],
               arr[n - 1] * arr[n - 2] * arr[n - 3])

if __name__ == "__main__":
    arr = [-10, -3, 5, 6, -20]
    result = maxProduct(arr)
    print(result)
```
