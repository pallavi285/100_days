# Principal Component Analysis 
We will look at a toy dataset first before moving on to more complex dataset. We will  now start using functions wherever applicable so that our code is more modular and reuseable.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rng = np.random.default_rng(seed=42)
```

# Toy Dataset

## Generate Dataset 
Let us first generate da dataset of R^2 by sampling points from a multivariate normal distribution .

```python
# Generate Dataset 
def generate(mu,cov,n):
    X = rng.multivariate_normal(mu,cov,n).T
    return X

mu = np.array([2,5])
cov = np.array([[1,0.9],
                [0.9,1]]) 
n = 100
X = generate(mu,cov,n)
print(X.shape)
```

# Visualise 
Let us now visualise the dataset using a scatter plot.

```python
def plot(X):
    plt.scatter(X[0,:],X[1,:],alpha=0.5)
    #alpha control opacity of the points 
    plt.axhline(color='black')
    plt.axvline(color='black')
    plt.axis('equal')
plot(X)
```

# Center the dataset 
Let us now center the dataset and visualise both the centerd and original dataset side by side.

```python
def center(X):
    mu = X.mean(axis =1)
    X -=mu.reshape(2,1)
    return X 

plt.subplot(1,2,1)
plot(X)
plt.title('Original')
X = center(X)
plt.subplot(1,2,2)
plot(X)
plt.title('centered')
```

# Covariance matrix 
Now ,we shall compute the covariance matrix 

```python
def covariance(X):
    d,n = X.shape 
    return X@X.T /n
C = covariance(X)
C
```
