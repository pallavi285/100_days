## np.zeros()
Creates array filled with 0

```python
import numpy as np
```

```python
# create 1D Array 
np.zeros(5)
#shape (5,)
```

```python
np.zeros((2,3))
```

### When to use
- Initializing matrices
- Machine learning weights
- placeholder arrays

## np.ones()
Create array filled with 1

```python
np.ones((3,2))
```

### When to use
- Bias terms in ML
- Normalization tasks
- testing formulas

## np.full()
Create array with same custom value

```python
np.full((2,2),10)
```

## np.eye()
Ones on specified diagonal and zeros elsewhere 
**(Identity matrix)**

```python
# 3X3 identity matrix 
print(np.eye(3))
```

```python
# 3X4 matrix with ones on main diagonal
print(np.eye(3,4))
```

```python
#shift diagonal up by 1
print(np.eye(4,k=1))
```

```python
#shift diagonal down by 2 
print(np.eye(5,k=-2))
```
