# rotation from scratch in deque

```python
from collections import deque
```

```python
def rotate_deque(dq,steps):
    if not dq or steps==0:
        return dq
    
    n = len(dq)
    steps = steps %  n #normalize
    
    # right rotation 
    if steps>0:
        for _ in range(steps):
            dq.appendleft(dq.pop())
            
    # left rotation
    else:
        for _ in range(abs(steps)):
            dq.append(dq.popleft())
    
    return dq
```

```python
dq = deque([1,2,3,4,5])
rotate_deque(dq,2)
print(dq)
```

```python
dq = deque([1,2,3,4,5])
rotate_deque(dq,5)
print(dq)
```

```python
dq = deque([1,2,3,4,5])
rotate_deque(dq,12)
print(dq)
```
