Given an array , return the index of the first element that repeats.   
if no element repeats , return -1.

Input : [10,5,3,4,5,6,7,5] (5(index 1 ) is the first element to repeat (index 4 ))\
Output : 5

```python
arr = [10,5,3,4,5,6,7,5]
seen = set()
ans = -1 

for i in range(len(arr)):
    if arr[i] in seen:
        ans = i        
    seen.add(arr[i])
print(ans)
```
