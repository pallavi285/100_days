# rotation in list

we use slicing to do this easily.

1Ô∏è‚É£ Left Rotation

üëâ Elements move towards left, first elements go to the end.

Example
lst = [1, 2, 3, 4, 5]
k = 2
rotated = lst[k:] + lst[:k]
print(rotated)

```python
# Take first k elements
# Put them at the end
lst = [1,2,3,4,5]
k=2 
rotated = lst[k:]+lst[:k]
print(rotated)
```

üß† Memory trick:
left ‚Üí cut first k ‚Üí paste at end

2Ô∏è‚É£ Right Rotation

üëâ Elements move towards right, last elements come to the front.

```python
lst = [1, 2, 3, 4, 5]
k = 2
rotated = lst[-k:] + lst[:-k]
print(rotated)
```

üß† Memory trick:
right ‚Üí cut last k ‚Üí paste at front
