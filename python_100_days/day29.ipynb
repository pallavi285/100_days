Abstract Class â†’ A blueprint or plan for other classes.
(You canâ€™t use it directly.)

Abstract Method â†’ A method without code â€” just a name.
(Child classes must fill it in later.)

Subclass â†’ The child class that uses the plan and writes the real code.

Object â†’ The final thing we create and use from the subclass.

```python
from abc import ABC, abstractmethod

#  Abstract Class (blueprint)
class Animal(ABC):
    
    # Abstract Method (task to complete later)
    @abstractmethod
    def sound(self):
        pass

# Subclass (child class that completes the task)
class Dog(Animal):
    def sound(self):
        print(" Dog says: Woof")

class Cat(Animal):
    def sound(self):
        print(" Cat says: Meow")

# Object (final usable form)
dog = Dog()
cat = Cat()

dog.sound()
cat.sound()
```
