# break()

`break` is used       
inside loops to       
**stop the loop**       
immediately

```python
for num in range(1, 10):
    if num == 5:
        break
    print(num)
    
# when the number reaches 5, the break statement stops the loop completely.
# So the output is 1, 2, 3, and 4 — and the loop ends.
```

Use `break` when you found what you're looking for, or when there’s no need to continue looping
