# Queue

## First In , First Out
- A Queue is a linear data structure.

![alt text](OIP.webp)

Front  ←  [ 10 | 20 | 30 ]  →  Rear

```python
from collections import deque 

qu = deque()
```

```python
#enqueue
qu.append(10)
qu.append(20)
# add at rear
print(qu)
```

```python
# dequeue
#remove from front
qu.popleft() 
print(qu)
```
