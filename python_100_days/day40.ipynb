# File handling

**File objects maintain an internal cursor (file pointer).**


- .read() reads the whole file and moves the pointer to EOF.

- .readline() reads one line and moves the pointer to the start of the next line.

- If the pointer is at EOF, further reads return an empty string.

- open(filename , mode)
- file pointer

```python
file = open("D:/learn/100_day\python_100_days/file_handling.txt")
file.read()
```

```python
line1 = file.readline()
line1
```

```python
lines = file.readlines() 
# all lines in list
print(lines)
```

```python
file.seek(0)
lines = file.readlines() 
# all lines in list

for i in lines:
    print(i)
```
