# module

- A module is a Python file that you can **import**.

- Helps keep **code clean and reusable**.

- Python has many **built-in modules**.

- You can **create your own modules** too.

```python
import math
print(math.sqrt(25))   # 5.0
```

**Built-in modules** (already available in Python)

Examples:

- `math`
- `random`
- `datetime`
- `os`
