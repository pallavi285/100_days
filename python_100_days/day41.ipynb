## seek()

- The seek() function is used to move the file cursor to a specific position inside a file.

```python
f = open("D:/learn/100_day\python_100_days/logfile.txt", "r")
print(f.read())
```

```python
f.seek(0)
```

```python
f.tell() #EOF
```

## tell()
- The tell() function in Python is used to determine the current position of the file pointer within a file

```python
f = open("D:/learn/100_day\python_100_days/logfile.txt", "r")
print(f.seek(2))
print(f.tell())
print(f.read())
f.close()
```
