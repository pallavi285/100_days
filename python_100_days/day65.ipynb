```python
import numpy as np
```

```python
# Single random interger
print(np.random.randint(10))
#(0 to 9)
# ka beach ma kuch bhi
```

```python
# random intergers in range 
print(np.random.randint(5,15))
# generates a single random
# integer between
# 5 (included) and 15 (excluded)
```
