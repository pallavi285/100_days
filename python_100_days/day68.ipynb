# numpy.mean()

 Compute the **arithmetic mean** (*average*) along the specified axis.

- numpy.mean(a , axis=None , dtype=None)
- **a** : Input array or list.
- **axis** : (None or int or tuples of ints , optional)
    - None : means entire array (defult)
    - 0 : means along columns 
    - 1 : mean along rows 
- **dtype** (optional)
    - dtype for computation 

```python
import numpy as np
```

```python
# 1D array 
arr1 = np.array([1,2,3,4,5])
print(np.mean(arr1))
```

```python
#2D array 
arr2 = np.array(
    [
        [1,2,3],
        [4,5,6]
    ]
)
print(arr2)
```

```python
print("mean of entire array:")
print(np.mean(arr2))
```

```python
# total columns = 3 
#[1,2,3]
#[4,5,6]

print("mean along columns: ")
print(np.mean(arr2,axis=0))
```

```python
# total row = 2 

#[1,2,3]
#[4,5,6]

print("mean along rows:")
print(np.mean(arr2,axis=1))
```
