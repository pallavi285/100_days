
# Arrays  in DSA and Python lists‚Ä¶ same or different? 


**‚ÄúIn DSA, an array is a straight block of memory ‚Äî elements stored one after another, like bricks in a line.‚Äù**

üß±üß±üß±üß±

‚ÄúThis makes access super fast.‚Äù


‚ÄúIn Python, lists behave like dynamic arrays ‚Äî they‚Äôre also stored in order, but they can auto-resize when you append.‚Äù

‚ÄúSo memory is still contiguous, just flexible.‚Äù


```python
import sys

a = []
print("Size of empty list:", sys.getsizeof(a))     # empty list

a.append(1)
print("Size:", sys.getsizeof(a))     # after adding 1 element

a.append(2)
print("Size:", sys.getsizeof(a))     # after adding 2nd element



print("using for loop for appending element")
for i in range(3,20):
    a.append(i)
    print(f"size:{ sys.getsizeof(a)}")
    
print(f"a:{a}")
```

Difference:\
`DSA arrays = fixed size`\
`Python lists = arrays that can resize`
