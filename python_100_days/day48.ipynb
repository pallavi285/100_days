# pass()

'pass' literally **does nothing**.
We use it as a **placeholder** when       
we don’t want to write code yet,     
but we still need the structure to be valid.

```python
for i in range(5):
    if i == 2:
        pass
    print(i)
```

Here, when i becomes 2, the pass statement runs but nothing happens — the loop continues normally.

Use pass in empty loops, empty classes, or functions you plan to complete later.
