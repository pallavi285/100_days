# What is Unicode encoding?


- Unicode is a **universal character set**     
that assigns a unique number (code point)    
 to     
 every  
 character    
 in every    
 language,   
  symbol, or   
   emoji.
- A code point     
 U+XXXX (e.g., U+0041 for 'A').     

 
- In Python 3, all str objects are sequences of Unicode characters — meaning Python can handle text from any language without special handling.

So Unicode has encoding formats:
UTF (Unicode Transformation Format)

✔ UTF-8 (most common)\
✔ UTF-16\
✔ UTF-32

## The **ord() function** in Python returns the Unicode code point of a given single character.

```python
decimal = ord('1')
print(decimal)
hex = f"U+{decimal:04X}"
hex
```


The **chr() function** is used to     
convert a Unicode code point (an integer) into    
its corresponding character (string of length 1).

```python
print(chr(65))     # Decimal code point
print(chr(0x41))   # Hexadecimal code point
```
