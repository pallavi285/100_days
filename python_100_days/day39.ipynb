# File handling

| Mode   | Meaning       | Use                                     |
| ------ | ------------- | --------------------------------------- |
| `"r"`  | Read          | To read existing file                   |
| `"w"`  | Write         | Creates new file or overwrites existing |
| `"a"`  | Append        | Adds content to end of file             |
| `"x"`  | Create        | Creates new file, error if exists       |
| `"r+"` | Read + Write  | No overwrite                            |
| `"w+"` | Write + Read  | Overwrites                              |
| `"a+"` | Append + Read | Cursor at end                           |
