# deque (double-ended queue)

### A **flexible line** where people can enter or leave from front or back.

#### ceating a deque

```python
from collections import deque

dq = deque(["pallavi", "krishna", "anamika"])
print(dq)
```

```python
type(dq)
```

```python
# adding element
dq.append(4)        # add to right
dq.appendleft(0)    # add to left
```

```python
dq
```

```python
# remove element
dq.pop()        # remove from right
dq.popleft()    # remove from left
```

```python
dq
```

```python
# peek element
print(dq[0])     # front element
print(dq[-1])    # last element
```
