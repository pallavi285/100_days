# Math module

```python
import math
```

## math.sqrt(x)

Returns the square root.

```python
print(math.sqrt(100))
```

## math.pow(x, y)

Returns $x^y$ as float.

```python
print(math.pow(2,3))
```

## math.floor(x)

Rounds **down** to nearest integer.

```python
print(math.floor(4.9))
```

## math.ceil(x)

Rounds **up** to nearest integer.

```python
print(math.ceil(5.1))
```
