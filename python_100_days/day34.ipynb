# The map() function

The **map()** function in Python \
***applies* a specified function** \
to \
each item of an iterable (like a list)\
and\
returns an **iterator**.\
You can convert this iterator to a list using the list() function.

```python
def square(n):
    return n**2

number = [1,2,3,4,5]
squared = map(square , number)
print(squared)
```

```python
print(list(squared))
```

How Map Objects Relate to Iterators\
**Iterator**: An object that implements `the __iter__()` and `__next__()` methods.\
 You can retrieve elements one at a time using next().\
**Map Object**: A specific type of iterator that applies a function lazily, which means elements are computed only when requested.
