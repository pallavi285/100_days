- This conversion is necessary because computers store and transmit data as bytes, not text.

**ASCII** (*American Standard Code for Information Interchange*)    
is a **character encoding standard**
used to represent text in computers.

## 1. ASCII is just a mapping between numbers and characters

ASCII defines a table of 128 values (from 0 to 127).

Each value represents one character.

Example:

65 → 'A'

97 → 'a'

48 → '0'

## 2. ASCII is strictly 7-bit

A single ASCII character uses exactly 7 bits.

Why? Because early computers stored data in 7-bit telecommunication systems.

7 bits = 2⁷ = 128 possible values.

## 3. ASCII cannot represent any language except English

It contains:

English A–Z, a–z

Digits

Basic punctuation

Control characters

NO Hindi, NO emojis, NO accented letters (é), NOTHING else.

## 4. ASCII’s first 32 values (0–31) are not even printable

These are control characters from old teletypes:

10 → Line Feed (LF)

13 → Carriage Return (CR)

9 → Horizontal Tab (TAB)

7 → Bell (made a physical bell ring)

These were hardware instructions, not text.

## 5. ASCII became the foundation of modern encodings

UTF-8 is designed so the first 128 code points are identical to ASCII.

Every UTF-8 file containing only English letters = valid ASCII file too.
