# Stack

## Last In First Out
- the last element added is the first one to be removed

![alt text](1.webp)

- Push(x) → add element to top

- Pop() → remove top element

- Peek / Top → see top element



```python
stack = []

stack.append(10)
stack.append(20)
stack.append(30) #add element

stack
```

```python
stack.pop()
```

```python
stack
```

```python
#peek see top element
print(stack[-1])
```
