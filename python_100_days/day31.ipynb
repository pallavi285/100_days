#  ‚≠ê Basic Array Operations

## üîπ Access (O(1))

```python
arr = [4,8,12,16]
print(arr[2]) #12
```

## üîπ Update (O(1))

```python
arr[1] = 20
arr
```

## üîπ Insert at end (Amortized) (O(n))

```python
arr.append(50)
arr
```

## üîπ Insert at index (O(n))

```python
arr.insert(3,99) # insert at index 3
arr
```

## üîπ Delete (O(n))

```python
arr.pop(1) # remove by index
arr
```
