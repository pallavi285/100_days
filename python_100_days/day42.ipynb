# with() statement
-  The with statement is used to      
ensure **proper acquisition** and      
**release of resources** by       
leveraging **context managers**.      
It *automatically* *handles* **setup** and **cleanup**,         
*even if exceptions occur*,       
making **code cleaner** and       
**safer** than using manual try...finally blocks.

```python
f = open("D:/learn/100_day/python_100_days/with",'a')
f.write('6.relese data')
f.close()
```

```python
# using with statement

with open("D:/learn/100_day/python_100_days/with",'a') as f:
    f.write('\n7.context managers')

# File is automatically closed here
```
