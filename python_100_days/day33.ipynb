# Python List Slicing

**Slicing** lets you extract a part of a list using:

*list*[**start** : **stop** : **step**]

- start → index where slice begins (included)

- stop → index where slice ends (excluded)

- step → how much to jump

For this list:
a = [10, 20, 30, 40, 50, 60]

```python
a = [10, 20, 30, 40, 50, 60]
```

| Value   | 10 | 20 | 30 | 40 | 50 | 60 |
|---------|----|----|----|----|----|----|
| Index   | 0  | 1  | 2  | 3  | 4  | 5  |
| -Index  | -6 | -5 | -4 | -3 | -2 | -1 |


```python
# 1️⃣ Basic Slice

a[1:4] 
# → elements from index 1 to 3
# Output: [20, 30, 40]
```

```python
# 2️⃣ Slice from beginning

a[:3] 
# → first 3 elements
# Output: [10, 20, 30]
```

```python
# 3️⃣ Slice till the end

a[2:]
# Output: [30, 40, 50, 60]
```

```python
# 5️⃣ Skip elements (step)

a[::2] 
# → every 2nd element
# Output: [10, 30, 50]
```

```python
# 6️⃣ Reverse the list

a[::-1]
# Output: [60, 50, 40, 30, 20, 10]
```

```python
# 7️⃣ Negative index slicing

a[-3:]
# Output: [40, 50, 60]
```
