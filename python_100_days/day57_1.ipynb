# rotation  and reverse in deque

**Rotation means:**

- Right rotation → take items from end → move to front (postive steps)

- Left rotation → take items from front → move to end (negative steps)

```python
from collections import deque
```

```python
# Right rotation
# → take items from end
# → move to front (postive steps)
dq = deque([1, 2, 3, 4])
dq.rotate(1)
print(dq)   # deque([4, 1, 2, 3]) 
```

```python
# Right rotation
# → take items from end
# → move to front (postive steps)
dq = deque([1, 2, 3, 4])
dq.rotate(2)
print(dq) 
```

```python
#Left rotation → 
# take items from front → 
# move to end (negative steps)
let = deque([1,2,3,4,5])
let.rotate(-1)
print(let)
```

```python
# Left rotation →
# take items from front →
# move to end (negative steps)
let = deque([1,2,3,4,5])
let.rotate(-3)
print(let)
```

```python
let = deque([1,2,3,4,5])
let.reverse()
print(let)
```
