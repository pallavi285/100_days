Polymorphism

**Polymorphism** = “Many forms.”\
It means the same function or method behaves differently depending on the object that calls it.

```python
class Bird:
    def sound(self):
        print("Chirp Chirp")

class Dog:
    def sound(self):
        print("Bark Bark")

# Same function name, different behaviors
for animal in (Bird(), Dog()):
    animal.sound()
```
