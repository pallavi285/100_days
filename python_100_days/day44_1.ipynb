# finally() 

### finally **runs ALWAYS** — success, error, or return. It handles the cleanup.

finally is used for **cleanup**:

Closing files,    
closing connections,    
releasing resources       
— things that must be done always,       
**even if an error came** or     
 **you returned early**.

```python
try:
   result = 10 / 0
   print(result)

finally:
   print("This block always executes.")
```

```python
def read_file(filename):
    file = None
    try:
        file = open(filename, "r")
        data = file.read()
        return data
    except FileNotFoundError:
        print("File not found!")
        return None
    finally:
        if file:
            file.close()
        print("File closed (finally block executed).")


# calling the function
print(read_file("sample.txt"))
# finally runs always wheather the file exist or not
```

- Can be used **without** an **except** block (try-finally).

- Ideal for **closing resources** and **ensuring exception-safe** code.
