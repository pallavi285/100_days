## 1. Using '=' for Copy (Assignment Copy)

= does NOT create a new copy.
It just creates another name for the same object.

```python
a = [1, 2, 3]
b = a
b[0] = 19
print(a)   # [99, 2, 3]
```

## 2. Shallow Copy (copy.copy())

Creates a new outer object

BUT inner nested objects still share memory

```python
import copy

a = [1, 2, 3]
b = copy.copy(a)

b[0] = 99
print(a)  # [1, 2, 3]
print(b)  # [99, 2, 3]

#work like deep copy becz 
#no nested object here
```

```python
import copy

a = [[1, 2], [3, 4]]
b = copy.copy(a)
b[0][0] = 99
print('a=',a)  
print('b=',b)     
# ‚ùó Change reflects in a
# ‚ùó Inner lists are shared
```

üîπ Objects where copy.copy() works well
immutable (cannot be changed)
‚úî List without nesting
‚úî Tuple with immutable items
‚úî Dictionary with immutable values
‚úî Set with immutable elements

## 3. Deep Copy (copy.deepcopy())

Creates a completely independent copy

Copies outer + all inner nested objects

```python
import copy

a = [[1, 2], [3, 4]]
b = copy.deepcopy(a)
b[0][0] = 99
print('a=',a)  
print('b=',b)  
```

üîπ When NOT to use Deep Copy

‚ùå Large data (slow & memory heavy)
‚ùå No nested objects
‚ùå Performance-critical code
